```python
import numpy as np
import random
import matplotlib.pyplot as plt

np.random.seed(10) #Stops from changing random values on every run of program

X = np.sort(np.random.uniform(low=0.0,high=1.0,size=500))
Y = np.sin(2*np.pi*X)
noise = np.random.normal(scale=0.3,size=Y.shape)
YY_ = Y + noise

plt.plot(X,YY_)
plt.show()
random.seed(10)
ind_= random.sample(range(500),500)
ind_train = ind_[:200]
X_sub_train = X[ind_train]
YY_sub_train = YY_[ind_train]
from sklearn.linear_model import LinearRegression
L=LinearRegression()
L.fit(X_sub_train[:,np.newaxis],YY_sub_train)
predictions=L.predict(X[:,np.newaxis])
#adding new axis means transforming array to matrix since predict take matrix only
plt.plot(X,Y,color='green')
plt.plot(X,predictions)
plt.show()
from sklearn.preprocessing import PolynomialFeatures
poly_ = PolynomialFeatures(degree=10)
X_sub_poly=poly_.fit_transform(X_sub_train[:,np.newaxis])
print(X_sub_poly.shape)
print(X_sub_poly)

X_poly=poly_.fit_transform(X[:,np.newaxis])
#to convert the input into 1,x,x^2,x^3
L.fit(X_sub_poly,YY_sub_train)
pred_ = L.predict(X_poly)
plt.plot(X,Y,color='green')
plt.plot(X,pred_,color='red')
plt.ylim(-1.25,1.25)
plt.scatter(X_sub_train,YY_sub_train,s=50,facecolors='none',edgecolors='b')
plt.show()
#more no. of data samples in dataset allows us to go for complex models .Complex model me less data samples leads to overfitting 
```
